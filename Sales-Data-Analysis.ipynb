```python
import pandas as pd
data = pd.read_csv("Sample - Superstore.csv",encoding="cp1252")
```

```python
data.head(5)
```

```python
data.tail(10)
```

```python
data.info()
```

**drop unnecessary column`**

```python
data.drop(columns=["Postal Code"])
```

**Convert Order Date to datetime**

```python
# Convert while allowing for format variations
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce', dayfirst=True)
data['Order Date'] = data['Order Date'].dt.strftime('%d/%m/%Y')
```

```python
data["Order Date"]
```

**Check and remove null values if needed**

```python
data.isnull().sum()
```

```python
data = data.dropna()
data.isnull().sum()
```

**Add columns for Month, Day, Year**

```python
#Convert Order Date to datetime
data["Order Date"] = pd.to_datetime(data["Order Date"])

#Extract Month, Day, Year
data["Month"] = data["Order Date"].dt.month
data["Day"] = data["Order Date"].dt.day
data["Year"] = data["Order Date"].dt.year

print(data[["Month","Day" ,"Year"]].head(5))
print()
print(data[["Month","Day" ,"Year"]].tail(5))
```

**Create a Total_Sale column**

```python
data["Total_sale"] = data["Sales"] - data["Sales"] * data["Discount"]
data[["Sales" , "Discount" , "Total_sale"]].head()
```

**Analysis**

**1 Top 5 products by sales**

```python
data.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(5)
```

**2. Most profitable cities**

```python
data.groupby("City")["Profit"].sum().sort_values(ascending=False).head(5)
```

**3. Monthly Sales**

```python
data.groupby("Month")["Sales"].sum()
```

**4. Average discount given**

```python
data["Discount"].mean()
```
